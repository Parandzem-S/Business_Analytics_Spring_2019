The <a> <a href= "https://jupyter.org/"> Jupyter Notebook</a> is used to produce our graphs by using Matplotlib library. Before starting with the first graph, we should do the following:

- The data of your interest should be read using <code>pandas.read()</code> function. For this blog post, Women's NBA players' on-court performance (2016-2017 season) data is used. The data is taken from <a><a href= "https://www.kaggle.com/jinxbe/wnba-player-stats-2017/home">www.kaggle.com</a>.

```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#reading data
my_data=pd.read_csv("https://raw.githubusercontent.com/Parandzem-S/Blog_metric.am/master/Data/WNBA%20Stats.csv")
```

<h2>Distribution Charts, Matplotlib</h2> <a name="distribution"></a>

<b> Frequency Histogram</b> <a name="histogram"></a>

A frequency histogram shows the frequency distribution of the data, i.e. how many times an outcome occurs in a data. It allows to organize data and not only to see the distribution type of data (e.g. normal distribution) but also to observe skewness, outliers, etc. <br>

Let's visualise the distribution of Women NBA (WMBA) players' age. 

```python
plt.figure(figsize=(8,6)) #making the plot bigger
#draw plot
plt.hist(my_data.Age, 
         bins=10, #defining number of bins
         color="#58508d" #changing color
)
plt.title("2016-2017 WNBA players' age distribution") #giving a title to the graph
plt.ylabel("Frequencies")#giving title to y axis
plt.xlabel("Age")#giving title to x axis
plt.show()#showing the plot
```

<b> Box Plot</b> <a name="box_plot"></a>

The another type of distribution chart graph is a box plot. Box plot uses five measures to describe how the data is distributed: <i>minimum, first quartile, median, third quartile, and maximum</i>. It is generally very useful tool to detect outliers in the data.  <br>

Let's visualise the distribution of WMBA players' height. 

```python
plt.figure(figsize=(8,6),facecolor="#e8eaed")#making the plot bigger
#draw plot
plt.boxplot(my_data.Height,
            patch_artist=True,  #fill boxes with color
            notch=True  #changing a rectangular form to a notched form
)
plt.annotate("%0.0F"% np.median(my_data.Height), xy=(0.98, 186), xytext=(0.98,186)) #annotating median value
#annotating names of 5 measures
plt.annotate('Minimum', xy=(1, 165), xytext=(1.1,164))
plt.annotate('First Quartile', xy=(1, 176), xytext=(1.1,176))
plt.annotate('Median', xy=(1, 184), xytext=(1.1,184))
plt.annotate('Third Quartile', xy=(1, 190.2), xytext=(1.1,190.2))
plt.annotate('Maximum', xy=(1, 205.5), xytext=(1.1,205.5))
plt.title("WNBA's Players Height (cm) Distribution") #giving a title to the plot
plt.grid(b=True) #showing grids
plt.show()
```

<h2>Relationship Charts, Matplotlib</h2> <a name="relate"></a>

<b> Scatter Plot</b> <a name="scatter"></a>

Scatter plots are used to visualize the relationship between two variables by using dots to plot the values of two variables along the x-axis and the y-axis. </br>

Let's visualise WNBA players' height and weight relationship using scatter plot. 

```python
plt.figure(figsize=(8,6), facecolor="#f2f4f9") #making the plot bigger
#draw plot
plt.scatter(my_data.Height,
            my_data.Weight,
            color="#004c6d")
#giving titles to axes
plt.title("Relationship between Height and Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (cm)")
plt.grid(b=True,
         linestyle=':' #giving line style to grid lines
)
plt.show()
```

<b> Bubble Plot</b> <a name="bubble"></a>

If we want to visualize the relationship between three variables then bubble plots can be used. The values of two variables are presented along x-axis and the y-axis, and the values of the third variable represent the size of the dots.</br>

Let's visualise the relationship between WNBA players' height, weight and field goals made (FGM) bubble plots. 

```python
plt.figure(figsize=(9,7), facecolor="#f2f4f9") #making the plot bigger
plt.scatter(my_data.Height, #defining x variable
            my_data.Weight, #defining y variable
            my_data.FGM*10, #the values of 3rd variable were muliplied by 10 to see better the size differences of dots
            c="#5886a5", #changing color
            alpha=0.8, #transparency of dots
            edgecolors="#f2f4f9", 
            linewidth=1.5)
# Add titles (main and on axis)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (cm)")
plt.title("Relationship between WNBA players' height, weight and FGM")
plt.show()
```

<h2>Comparison Charts, Matplotlib</h2> <a name="compare"></a>

<b> Bar Plots</b><br> <a name="bar"></a>
A bar plot is a plot/chart that presents categorical data with rectangular bars with lengths or heights proportional to the values that they represent. One axis of horizontal or vertical bar plot shows the categories or groups being compared, and the other axis represents a measured value.

Let's observe the realtionship between player's position and their height by using bar plot.

```python
#Vertical bar plot
plt.figure(figsize=(9,7),facecolor="#f2f4f9")
#draw plot
plt.bar(my_data.Pos, my_data.Height, color="#004c6d")
plt.ylim(0,250) #define the range of Y axis
#giving titles
plt.title("WNBA Players Height by Position")
plt.ylabel("Height (cm)")
plt.xlabel("Positions")
plt.show()
```

```python
#Horizontal Bar plot
plt.figure(figsize=(9,7),facecolor="#f2f4f9")
#draw plot
plt.barh(my_data.Pos, my_data.Weight,color="#5886a5")
#giving titles
plt.title("WNBA Players Weight by Position")
plt.ylabel("Positions")
plt.xlabel("Weight (kg)")
plt.xlim(0,130) #define the range of X axis
plt.show()
```

<b> Line Plot </b> <a name="line"></a>

Let's use line plots to compare average height of WNBA players by their position.

```python
#calculating mean heights by position
mean_height=my_data.Height.groupby(my_data.Pos).mean() #returns the mean value by each position
posistion=my_data.Pos.unique() #returns array with names of positions

plt.figure(figsize=(9,7),facecolor="#f9f5f2") #making plot bigger
#draw plot
plt.plot(posistion, 
         mean_height, 
         color="#ff8019",
         marker='o')
#giving titles
plt.title("WNBA Players Mean Height by Position")
plt.ylabel("Height (cm)")
plt.xlabel("Positions")
plt.show()
```

<h2>Composition Charts, Matplotlib</h2><a name="comp"></a>

<b>Stacked Bar Plot</b> <a name="stacked"></a>

Simple Stacked Bar plots place each value for the category after the previous one. The total value of the bar is all the sum of values of all categories/groups. This graph is used to compare the total amounts across each category/group bar.

```python
#Vertical bar plot, use plt.barh() to drow Horizontal Stacked Bar Plot as well
plt.figure(figsize=(9,7),facecolor="#f2f4f9")
#draw plot
plt.bar(my_data.Pos, my_data.FGA,color="#5886a5", label='FGA')#label will allow to have names of variables on graph
plt.bar(my_data.Pos, my_data.FGM, color="#004c6d", label='FGM') #label will allow to have names of variables on graph
plt.ylim(0,800) #define the range of y axis
#giving titles
plt.title("Field Goals Made and Field Goals Attempts")
plt.xlabel("Positions")
plt.legend() #to create legend so that we can differenciate variables
plt.show() #show graph
```

<b>Pie Chart</b> <a name="pie"></a>

A pie chart presents numerical proportions(percentages) between categories by deviding a circle into slices. In a pie chart, the arc length of each slice (and consequently its central angle and area), is proportional to the quantity it represents, and the the full circle represents the total sum of all proportions. It can give a quick idea about proportional comparison of data. However, this type of chart is not used much because it is difficult to compare areas (or angels) if the difference is not evident.

Let's observe proportianl distribution of number of players by position.

```python
col=["#e8a668","#773f3f","#4f7b8a", "#d9f1f6","#89c1ec"] #define colors for each slice of pie chart
plt.figure(figsize=(8,8),facecolor="#f2f4f9") #change plot size
#draw plot
plt.pie(my_data.Pos.value_counts(),
        labels=my_data.Pos.unique(),
        colors=col)
plt.title("Players by position") #giving a title
plt.legend() #to create legend so that we can differenciate variables
plt.show()
```
